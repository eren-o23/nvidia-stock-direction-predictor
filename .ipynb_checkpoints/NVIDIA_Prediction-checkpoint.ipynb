```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score
)
```

```python
# Load Data
df = pd.read_csv('nvidia_stock.csv', parse_dates=['Date'])
df = df.sort_values('Date')
```

```python
# Create the target: 1 if tomorrow's Close is higher than today's
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)
```

```python
# Feature Engineering
df['Return'] = df['Close'].pct_change()
df['Vol_Change'] = df['Volume'].pct_change()
df['MA10'] = df['Close'].rolling(window=10).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()
df['Volatility'] = (df['High'] - df['Low']) / df['Close']
```

```python
# Define Features and Target
features = ['Return', 'Vol_Change', 'Volatility', 'MA10', 'MA50']
X = df[features]
y = df['Target']
```

```python
# Time-Series Split
split = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]
```

```python
# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models = {
    "Untuned LR": LogisticRegression(random_state=42),
    "Untuned RF": RandomForestClassifier(random_state=42),
    "Untuned GBT": GradientBoostingClassifier(random_state=42)
}

print("Baseline Performances")
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    acc = accuracy_score(y_test, model.predict(X_test_scaled))
    print(f"{name} Accuracy: {acc:.4f}")
```

```python
# 1. Tune Logistic Regression
param_lr = {'C': [0.1, 1, 10]}
grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_lr, cv=3).fit(X_train_scaled, y_train)
best_lr = grid_lr.best_estimator_

# 2. Tune Random Forest
param_rf = {'n_estimators': [100, 200], 'max_depth': [3, 5]}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_rf, cv=3).fit(X_train_scaled, y_train)
best_rf = grid_rf.best_estimator_

# 3. Tune GBT
param_gbt = {'learning_rate': [0.01, 0.1], 'n_estimators': [100]}
grid_gbt = GridSearchCV(GradientBoostingClassifier(random_state=42), param_gbt, cv=3).fit(X_train_scaled, y_train)
best_gbt = grid_gbt.best_estimator_
```

```python
# Create Tuned Ensemble
ensemble = VotingClassifier(
    estimators=[('lr', best_lr), ('rf', best_rf), ('gbt', best_gbt)],
    voting='soft'
)

# Train and Evaluate Ensemble
ensemble.fit(X_train_scaled, y_train)
y_pred_ensemble = ensemble.predict(X_test_scaled)

# Comparison Report
final_models = {
    "Tuned LR": best_lr,
    "Tuned RF": best_rf,
    "Tuned GBT": best_gbt,
    "Final Ensemble": ensemble
}

print("\n--- Final Model Comparison (Accuracy) ---")
for name, model in final_models.items():
    acc = accuracy_score(y_test, model.predict(X_test_scaled))
    print(f"{name}: {acc:.4f}")

print("\n--- Detailed Ensemble Report ---")
print(classification_report(y_test, y_pred_ensemble))
```
